```python
# 요 연산은, final layer의 softmax함수의 derivative 함수가 모든 layer에서 1이라는
# 가정하에 연산한것이고, 원래는 softmax함수의 derivative를 곱해야합니다.
# first_layer에서는 Relu함수를 구하기때문에, Relu에 대한 derivative를 곱할것입니다
delta_weight = np.matmul(error, multi1.transpose())

lr = 0.0001  # learning rate = 0.01
for i in range(NOUT):
    for j in range(NHIDDEN):
        # delta_weight[i, j] = int(delta_weight[i, j] * lr)
        delta_weight[i, j] = delta_weight[i, j] * lr
weight_memory2 = weight_memory2 - delta_weight

pd.DataFrame(delta_weight)
```

```python
delta_bias = error
lr = 0.0001 # learning rate = 0.005
for i in range(NOUT):
    # delta_bias[i, 0] = int(delta_bias[i, 0] * lr)
    delta_bias[i, 0] = delta_bias[i, 0] * lr
bias_memory2 = bias_memory2 - delta_bias

pd.DataFrame(delta_bias)
```

### model parameter definition

```python
import numpy as np     # numpy는 array data를 만들 수 있음
import pandas as pd    # pandas는 matrix data를 만들 수
from random import *  # random 변수

def relu(x):
    if x > 0:
        return x
    else:
        return 0
    

def one_hot(dec):
    value = np.zeros((10, 1))
    if (dec >= 0 and dec < 10):
        value[dec, 0] = 1
    return value

NHIDDEN = 256
NTEST = 60000
NDATA = 784
LOG1 = 10
LOG2 = 8
NOUT = 10

# mnist_test = pd.read_csv('./mnist_data/mnist_test.csv', sep=",", header = None)
mnist_train = pd.read_csv('./mnist_data/mnist_train.csv', sep=",", header = None)

mnist_label = mnist_train.iloc[:,0]
mnist_fixel = mnist_train.iloc[:,1:]

mnist_label = np.array(mnist_label)  # 1-dimensional vector
mnist_fixel = np.array(mnist_fixel)  # 2-dimensional array

weight_memory1 = np.zeros((NHIDDEN, NDATA))
weight_memory2 = np.zeros((NOUT, NHIDDEN))
bias_memory1 = np.zeros((NHIDDEN, 1))
bias_memory2 = np.zeros((NOUT, 1))

mu, sigma1, sigma2 = 0, 0.05, 0.09 # mean and standard deviation
# np.random.nomral 함수를 이용해서 평균 0, 표준편차 0.1인 sample들을 1000개 추출한다.
s1 = np.random.normal(mu, sigma1, NDATA*NHIDDEN)
s2 = np.random.normal(mu, sigma2, NHIDDEN*NOUT)

weight_memory1 = s1.reshape(NHIDDEN, NDATA)
weight_memory2 = s2.reshape(NOUT, NHIDDEN) 

```

```python
mnist_train = pd.read_csv('./mnist_data/mnist_train.csv', sep=",", header = None)

mnist_label = mnist_train.iloc[:,0]
mnist_fixel = mnist_train.iloc[:,1:]

mnist_label = np.array(mnist_label)  # 1-dimensional vector
mnist_fixel = np.array(mnist_fixel)  # 2-dimensional array
```

### model train (60K)

```python
NLOOP = 20
NTRAIN_DATA = 60000
lr = 0.00001

for loop in range(NLOOP):      # epoch 수
    for i in range(0, NTRAIN_DATA):  # 60K Data 학습임(test아님)
                                     # batch_size = 1  
        if i%10000 == 0: 
            print("{} is start".format( (i) / 10000))
        ## [1.1] Forward propagation executing for first layer
        input_data = mnist_fixel[i, :]  # fixel data인 784개를 먼저 가져옴
        input_data = input_data.reshape(NDATA, 1)
        multi1 = np.matmul(weight_memory1, input_data)  # weight multiply
        multi1 = multi1.reshape(NHIDDEN, 1)  # shape이 (256, )로 되는것을 막는다.

        multi1 = np.add(multi1, bias_memory1)  # adding bias

        # take relu function each element
        # multi1는 첫번째 hidden layer의 output이 됩니다.
        for j in range(NHIDDEN):
            multi1[j, 0] = relu(multi1[j, 0])

        
        ## [1.2] Forward propagation for second layer
        multi2 = np.matmul(weight_memory2, multi1)
        multi2 = np.add(multi2, bias_memory2)

        ## [1.3] find largest value of our calculation
        count = 0
        max = multi2[0, 0]
        for j in range(NOUT):
            if (max < multi2[j, 0]):
                max = multi2[j, 0]
                count = j
    
        # our expected one_hot vector
        expected = one_hot(count)
    
        # our label data
        label = one_hot(mnist_label[i])

        # [1.4] checking expected data is same for label
        flag = True
        for j in range(NOUT):
            if (expected[j, 0] != label[j, 0]):
                flag = False
        
        # if our expected value is correct, 
        # forward propagation another data(don't need back propagation)
        if (flag):
            continue

        # [2] back propagation process (expected data is mismatched)
        # this process supposed to derivative of softmax function is equal to 1
        # all of layer, originally we should multiply 
        # derivative of softmax function
        error = expected - label

        # [2.1] calculate delta_weight2 of weight_memory2
        delta_weight2 = np.matmul(error, multi1.transpose())
        # lr = 1/8192  # learning rate = 0.01
        for num1 in range(NOUT):
            for num2 in range(NHIDDEN):
                # delta_weight2[num1, num2] = int(delta_weight2[num1, num2] * lr)
                delta_weight2[num1, num2] = delta_weight2[num1, num2] * lr

        weight_memory2 = weight_memory2 - delta_weight2

        # [2.2] calculate delta_bias of bias_memory2
        delta_bias2 = error
        # lr = 1/8192
        for num1 in range(NOUT):
            # delta_bias2[num1, 0] = int(delta_bias2[num1, 0] * lr)
            delta_bias2[num1, 0] = delta_bias2[num1, 0] * lr
        bias_memory2 = bias_memory2 - delta_bias2

        # [2.3] derivative of layer unit(derivative of activation function = 1)
        neuron_act = np.matmul(weight_memory2.transpose(), error)
    
        # multi1 = first layer output(by taking relu function)
        # del_multi1 = derivative of relu function of multi1
        del_multi1 = np.zeros((NHIDDEN, 1))
        for num1 in range(NHIDDEN):
            if multi1[num1, 0] > 0:
                del_multi1[num1, 0] = 1
    
        # [2.4] neuron_act * del_multi1
        neuron_act_mul = np.zeros((NHIDDEN, 1))
        for num1 in range(NHIDDEN):
            neuron_act_mul[num1, 0] = del_multi1[num1, 0] * neuron_act[num1, 0]

        # [2.5] calculate delta_weight1 
        delta_weight1 = np.matmul(neuron_act_mul, input_data.transpose())
    
        # testing delta_weight1
        # if i % 50 == 0:
        #    Data = pd.DataFrame(delta_weight1)
        #    path = './delta_weight/delta_weight1_%d'%(i)   # path = '파일 경로/저장할 파일 이름'
        #    Data.to_csv(path, sep=',', header=None , index=None)

        # lr = 1/8192  # learning rate = 0.01
        for num1 in range(NHIDDEN):
            for num2 in range(NDATA):
                # delta_weight1[num1, num2] = int(delta_weight1[num1, num2] * lr)
                delta_weight1[num1, num2] = delta_weight1[num1, num2] * lr

        weight_memory1 = weight_memory1 - delta_weight1
    
        # [2.6] calculate delta_bias1
        # lr = 1/8192
        delta_bias = np.zeros((NHIDDEN, 1))

        for num1 in range(NHIDDEN):
            # delta_bias[i, 0] = int(neuron_act_mul[i, 0] * lr)
            delta_bias[num1, 0] = neuron_act_mul[num1, 0] * lr
      
        bias_memory1 = bias_memory1 - delta_bias
    
    
    print("loop {} is end!".format(loop))
```

```python
import numpy as np
class NumpyFloatToFixConverter(object):
    """A callable which converts Numpy arrays of floats to fixed point arrays.
    General usage is to create a new converter and then call this on arrays of
    values.  The `dtype` of the returned array is determined from the
    parameters passed.  For example::
        >>> f = NumpyFloatToFixConverter(signed=True, n_bits=8, n_frac=4)
    Will convert floating point values to 8-bit signed representations with 4
    fractional bits.  Consequently the returned `dtype` will be `int8`::
        >>> import numpy as np
        >>> vals = np.array([0.0, 0.25, 0.5, -0.5, -0.25])
        >>> f(vals)
        array([ 0,  4,  8, -8, -4], dtype=int8)
    The conversion is saturating::
        >>> f(np.array([15.0, 16.0, -16.0, -17.0]))
        array([ 127,  127, -128, -128], dtype=int8)
    The byte representation can be expected to match that for using
    `float_to_fix`::
        >>> d = f(np.array([-16.0]))
        >>> import struct
        >>> g = float_to_fix(True, 8, 4)
        >>> val = g(-16.0)
        >>> struct.pack('B', val) == bytes(d.data)
        True
    An exception is raised if the number of bits specified cannot be
    represented using a whole `dtype`::
        >>> NumpyFloatToFixConverter(True, 12, 0)
        Traceback (most recent call last):
        ValueError: n_bits: 12: Must be 8, 16, 32 or 64.
    """
    dtypes = {
        (False, 8): np.uint8,
        (True, 8): np.int8,
        (False, 16): np.uint16,
        (True, 16): np.int16,
        (False, 32): np.uint32,
        (True, 32): np.int32,
        (False, 64): np.uint64,
        (True, 64): np.int64,
    }

    def __init__(self, signed, n_bits, n_frac):
        """Create a new converter from floats into ints.
        Parameters
        ----------
        signed : bool
            Indicates that the converted values are to be signed or otherwise.
        n_bits : int
            The number of bits each value will use overall (must be 8, 16, 32,
            or 64).
        n_frac : int
            The number of fractional bits.
        """
        # Check the number of bits is sane
        if n_bits not in [8, 16, 32, 64]:
            raise ValueError(
                "n_bits: {}: Must be 8, 16, 32 or 64.".format(n_bits))

        # Determine the maximum and minimum values after conversion
        if signed:
            self.max_value = 2**(n_bits - 1) - 1
            self.min_value = -self.max_value - 1
        else:
            self.max_value = 2**n_bits - 1
            self.min_value = 0

        # Store the settings
        self.bytes_per_element = n_bits / 8
        self.dtype = self.dtypes[(signed, n_bits)]
        self.n_frac = n_frac

    def __call__(self, values):
        """Convert the given NumPy array of values into fixed point format."""
        # Scale and cast to appropriate int types
        vals = values * 2.0 ** self.n_frac

        # Saturate the values
        vals = np.clip(vals, self.min_value, self.max_value)

        # **NOTE** for some reason just casting resulted in shape
        # being zeroed on some indeterminate selection of OSes,
        # architectures, Python and Numpy versions"
        return np.array(vals, copy=True, dtype=self.dtype)
    
conv = NumpyFloatToFixConverter(signed=True, n_bits=16, n_frac=14)

1.4 - (-1.2) 
```
